# Plot Coherence vs Lidar Height

```python
import os
import numpy as np
import gdal
import matplotlib.pyplot as plt
```

```python
home_dir = os.path.join(os.getenv("HOME"), "geodata", "geo411", "GEO411_FSH_Roda")
# change to home_dir
os.chdir(home_dir)
```

# Define Paths

```python
# uncropped, original resolution, original crs dem
dem = "ancillary_data/LAS_diff_Roda_10m_median_NA.tif"

# uncropped, vrt, wrong projection (no), wrong resolution
coh_hv =  "20150821_20151002/interferogram/topophase.cor.geo.vrt"
```

# Set NoData Value in Lidar DEM to -99!!

```python
dem_new = "ancillary_data/LAS_diff_Roda_10m_nodata99_median_NA.tif"
if not os.path.exists(dem_new):
    cmd = "gdal_translate -a_nodata -99 {dem_old} {dem_new}".format(dem_old=dem, dem_new=dem_new)
    os.system(cmd)
else: print("DEM with new no data value already exists!")
```

# Make vrt a GeoTiff

```python
# this is important as it converts the correlation-vrt to a GeoTiff
coherence_tif = "20150821_20151002/interferogram/topophase.cor.geo.tif"

if not os.path.exists(coherence_tif):
    # make reprojected, GeoTiff of band 2
    # from now on coherence has only one band
    cmd = "gdal_translate -of GTiff -b 2 {coh_hv} {tiff}".format(coh_hv = coh_hv,
                                                                tiff=coherence_tif)
    os.system(cmd)
else: print("Coherence already exists in GeoTiff")
```

# Reproject Lidar DEM and Coherence into UTM

```python
# check spatial references
print("COHERENCE SRS")
!gdalsrsinfo -o proj4 "20150821_20151002/interferogram/topophase.cor.geo.tif"
print("DEM SRS")
!gdalsrsinfo -o proj4 "ancillary_data/LAS_diff_Roda_10m_int16_median_NA.tif"
```

```python
reprojected_coh = "20150821_20151002/interferogram/topophase.cor.geo_32632.tif"
reprojected_dem = "ancillary_data/roda_10m_nodata99_32632.tif"


if not os.path.exists(reprojected_dem):
    # reproject dem
    cmd = "gdalwarp -t_srs EPSG:32632 {dem} {dem_repro}".format(dem=dem_new, dem_repro=reprojected_dem)
    os.system(cmd)
else: print("10m Lidar DEM already exists in 32632")

if not os.path.exists(reprojected_coh):
    # reproject coh
    cmd = "gdalwarp -t_srs EPSG:32632 {coh_tif} {coh_repro}".format(coh_tif=coherence_tif,
                                                                   coh_repro=reprojected_coh)
    os.system(cmd)

else: print("Reprojected Coherence already exists")
```

```python
ds1 = gdal.Open(reprojected_coh)
prj_coh = ds1.GetProjection() 
ds2 = gdal.Open(reprojected_dem)
prj_dem = ds2.GetProjection()
# check if both datasets have exactly the same projection definition
print(prj_coh == prj_dem)
```

# Compute Tileindexes


```python
shape_path = "20150821_20151002/shapes"
if not os.path.exists(shape_path):
    os.makedirs(shape_path)
else: print("Directory for shapes already exists")
```

```python
tile_coh = "20150821_20151002/shapes/extent.shp"
tile_dem = "ancillary_data/las_10m_extent.shp"

if not os.path.exists(tile_coh):
    cmd = "gdaltindex {coh_tile} {coh_tif}".format(coh_tile=tile_coh, 
                                                    coh_tif=reprojected_coh)
    os.system(cmd)
else: print("Tile for reprojected backscatter SAR Scene already exists")

if not os.path.exists(tile_dem):
    cmd = "gdaltindex {dem_extent_shape} {dem}".format(dem_extent_shape=tile_dem, 
                                                    dem=reprojected_dem)
    os.system(cmd)
else: print("Tile for reprojected DEM already exists")
```

# Calculate Intersection
- would be possible from the command line with ogr
- However, lets use geopandas

```python
import geopandas as gpd

# read lidar extent
lidar_extent = gpd.read_file(tile_dem)
lidar_extent["class"] = "Extent Lidar Dem"
# read coh extent
coh_extent = gpd.read_file(tile_coh)
coh_extent["class"] = "Extent Coherence"
```

```python
# stack dataframes (needs normal pandas...)
import pandas as pd
df = gpd.GeoDataFrame(pd.concat([lidar_extent, coh_extent]))
```

```python
df.plot(alpha=.7, column="class", legend=True)
```

# Compute Intersection

```python
intersection_shape = "20150821_20151002/shapes/intersection.shp"

if not os.path.exists(intersection_shape):
    intersection = gpd.overlay(lidar_extent, coh_extent, how="intersection")
    intersection.to_file(intersection_shape)
else: print("intersection already exists")
```

# Crop both rasters to intersection shape

```python
cropped_reprojected_20m_dem = "ancillary_data/roda_20m_32632_nodata99_cropped.tif"
cropped_reprojected_20m_coh = "20150821_20151002/interferogram/cropped_topophase_cor_geo_20m_srcnodata0.tif"

if not os.path.exists(cropped_reprojected_20m_dem):
    cmd = "gdalwarp -cutline {intersection} -crop_to_cutline -tr 20 20  {dem} {dem_cropped}".format(intersection=intersection_shape,
                                                                                                   dem=reprojected_dem,
                                                                                                   dem_cropped=cropped_reprojected_20m_dem)
    os.system(cmd)

else: print("cropped, reprojected and resampled DEM already exists")
    
if not os.path.exists(cropped_reprojected_20m_coh):
    cmd = "gdalwarp -cutline {intersection} -crop_to_cutline -tr 20 20 -srcnodata 0 {coh} {coh_cropped}".format(intersection=intersection_shape,
                                                                                                   coh=reprojected_coh,
                                                                                                   coh_cropped=cropped_reprojected_20m_coh)
    os.system(cmd)
else: print("cropped, reprojected and resampled sar-backsactter already exists")
```

# ALL VALUES BELOW 0 are NA in LIDAR DEM
- the handling of nodata values can be a mess
- gdal and numpy are somewhat different
- use ure common sense and set all values < 0 to `np.nan`

```python
ds_dem = gdal.Open(cropped_reprojected_20m_dem, gdal.GA_ReadOnly)
dem = ds_dem.GetRasterBand(1)
dem_nodata = dem.GetNoDataValue()
dem_array = dem.ReadAsArray()
dem_array[dem_array < 0] = np.nan
plt.imshow(dem_array)
plt.colorbar()
```

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable
import matplotlib 

# read both rasters with gdal
ds_coh = gdal.Open(cropped_reprojected_20m_coh, gdal.GA_ReadOnly)
ds_dem = gdal.Open(cropped_reprojected_20m_dem, gdal.GA_ReadOnly)

# get the first band 
coh = ds_coh.GetRasterBand(1)
dem = ds_dem.GetRasterBand(1)

# no-data values
dem_nodata = dem.GetNoDataValue()
coh_nodata = coh.GetNoDataValue()

# Snap can't handle nodatavals in export of geotiff really well...
print("[ NO DATA VALUE Coherence] = ", coh.GetNoDataValue())
print("[ NO DATA VALUE Dem] = ", dem.GetNoDataValue())




# read them into an array
coh_array = coh.ReadAsArray()
dem_array = dem.ReadAsArray()

# set -99 in lidar dem to na
dem_array[dem_array < 0] = np.nan
# set 0?? in coherence to na
coh_array[coh_array == 0] = np.nan

fig, (axs1, axs2) = plt.subplots(1,2, figsize=(18,16))
# save the axis in a an object returned by axs1.imshow
coh = axs1.imshow(coh_array, cmap="viridis")
divider = make_axes_locatable(axs1)
cax1 = divider.append_axes("right", size="5%", pad=.05)
fig.colorbar(coh, cax=cax1)
axs1.set_title("Coherence")


dem = axs2.imshow(dem_array, cmap="viridis")
divider = make_axes_locatable(axs2)
cax2 = divider.append_axes("right", size="5%", pad=.05)
fig.colorbar(dem, cax=cax2)
axs2.set_title("Lidar DEM in m")

plt.tight_layout(h_pad=1)
```

```python
# count nas
print("There are {} NAs in the Lidar DEM".format(np.count_nonzero(np.isnan(dem_array))))
print("There are {} NAs in the Coherence Tiff".format(np.count_nonzero(np.isnan(coh_array))))
```

```python
# something is slightly off with the array dimensions
print(dem_array.shape)
print(coh_array.shape)
```

```python
# set equally NAs
coh_array[np.where(np.isnan(dem_array))] = np.nan
dem_array[np.where(np.isnan(coh_array))] = np.nan

fig, (axs1, axs2) = plt.subplots(1,2, figsize=(18,16))
axs1.imshow(coh_array, cmap="gray")
axs2.imshow(dem_array, cmap="gray")
```
