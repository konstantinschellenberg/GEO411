# define paths

```python
# define path for dem and sar image
dem = "ancillary_data/LAS_diff_Roda_10m_median_NA.tif"
bs_hv =  "snap_grd/20150821/ALOS2-FBDR1_1__A-ORBIT__ALOS2067251007-150821_Cal_TF_TC_dB_hv.tif"
```

# Reproject data

```python
reprojected_dem = "ancillary_data/roda_10m_32632.tif"
reprojected_backscatter = "snap_grd/20150821/ALOS2_20150821_db_32632.tif"

if not os.path.exists(reprojected_dem):
    # reproject dem
    cmd = "gdalwarp -t_srs EPSG:32632 {dem_4326} {dem_repro}".format(dem_4326=dem, dem_repro=reprojected_dem)
    os.system(cmd)
else: print("10m Lidar DEM already exists in 32632")

if not os.path.exists(reprojected_backscatter):
    # reproject backscatter
    cmd = "gdalwarp -t_srs EPSG:32632 {bs_4326} {bs_repro}".format(bs_4326=bs_hv, bs_repro=reprojected_backscatter)
    os.system(cmd)
else: print("The reprojected Backscatter already exists")
```

# Compute Tileindexes


```python
tile_sar = "ancillary_data/sar_extent.shp"
tile_dem = "ancillary_data/las_10m_extent.shp"

if not os.path.exists(tile_sar):
    cmd = "gdaltindex {sar_extent_shape} {sar_backscatter_tif}".format(sar_extent_shape=tile_sar, 
                                                                sar_backscatter_tif=reprojected_backscatter)
    os.system(cmd)
else: print("Tile for reprojected backscatter SAR Scene already exists")

if not os.path.exists(tile_dem):
    cmd = "gdaltindex {dem_extent_shape} {dem}".format(dem_extent_shape=tile_dem, 
                                                    dem=reprojected_dem)
    os.system(cmd)
else: print("Tile for reprojected DEM already exists")
```

# Crop both rasters to intersection

```python
cropped_reprojected_20m_dem = "ancillary_data/roda_20m_32632_cropped.tif"
cropped_reprojected_20m_sar = "snap_grd/20150821/ALOS2_20150821_db_hv_32632_20_20_cropped.tif"

if not os.path.exists(cropped_reprojected_20m_dem):
    cmd = "gdalwarp -cutline {intersection} -crop_to_cutline -tr 20 20 -dstnodata -99 {dem} {dem_cropped}".format(intersection=intersection_shape,
                                                                                                   dem=reprojected_dem,
                                                                                                   dem_cropped=cropped_reprojected_20m_dem)
    os.system(cmd)
else: print("cropped, reprojected and resampled DEM already exists")
    
if not os.path.exists(cropped_reprojected_20m_sar):
    cmd = "gdalwarp -cutline {intersection} -crop_to_cutline -tr 20 20 -dstnodata -99 {sar} {sar_cropped}".format(intersection=intersection_shape,
                                                                                                   sar=reprojected_backscatter,
                                                                                                   sar_cropped=cropped_reprojected_20m_sar)
    os.system(cmd)
else: print("cropped, reprojected and resampled sar-backsactter already exists")
```

```python
import numpy as np
import rasterio
from rasterio import features
from rasterio.mask import mask
import fiona
from geojson import Point, Feature, FeatureCollection, dump
```

```python
# read both rasters with gdal
ds_bs = gdal.Open(cropped_reprojected_20m_sar, gdal.GA_ReadOnly)
ds_dem = gdal.Open(cropped_reprojected_20m_dem, gdal.GA_ReadOnly)

# get the first band 
bs = ds_bs.GetRasterBand(1)
dem = ds_dem.GetRasterBand(1)

# Snap can't handle nodatavals in export of geotiff really well...
print("[ NO DATA VALUE ] = ", bs.GetNoDataValue())

# read them into an array
bs_array = bs.ReadAsArray()
dem_array = dem.ReadAsArray()

fig = plt.figure(figsize = (10,10))

ax = fig.add_subplot(1,2,1)
ax.imshow(bs_array)

ax = fig.add_subplot(1,2,2)
ax.imshow(dem_array)
```

```python
# check if they have the same extent
bs_array.shape == dem_array.shape
```

- as Snap doesn't export any NoData, the parts where the backscatter is 0 is assumed to be Nodata


```python
# set all items in dem-array no na, that are 0 in backscatter
dem_array[np.where(bs_array == 0)] = np.nan

# set all items in backscatter-array to na, that are -99 in dem-array
bs_array[np.where(dem_array == -99 )] = np.nan

# set all items in backscatter-array to na that are 0 in backscatter-array
bs_array[np.where(bs_array == 0)] = np.nan

# set all items in dem-array to na, that are -99 in dem-array
dem_array[np.where(dem_array == -99)] = np.nan

fig = plt.figure(figsize = (10,10))

ax = fig.add_subplot(1,2,1)
ax.imshow(bs_array)

ax = fig.add_subplot(1,2,2)
ax.imshow(dem_array)
```
